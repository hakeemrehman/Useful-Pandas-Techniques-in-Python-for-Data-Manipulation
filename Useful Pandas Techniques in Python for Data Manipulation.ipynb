### 2. Groupby Function and Pivot Table in Pandas

#### Groupby Function

```python
# How many time each particulcar country value repeated in data?
sales.groupby('Country').count() 
```

```python
# How many time each particulcar country value repeated in data?
sales.groupby('Country')['Country'].count()
```

```python
# How many time country + Age_Group wise values repeated in data?
sales.groupby(['Country', 'Age_Group'])['Country'].count()
```

```python
sales.groupby(['Country', 'Age_Group'])['Country'].count().unstack()
```

#### Groupby Function + Descriptive Analysis

```python
# Computing average profit country wise
sales.groupby('Country')['Profit'].mean()
```

```python
# Computing average of Profit and ordered quntity in particular month
sales.groupby('Month')[['Profit','Order_Quantity']].mean() 
```

```python
# Computing average of Profit and ordered quntity in particular month and Age_group wise
sales.groupby(['Month','Age_Group'])[['Profit','Order_Quantity']].mean()  
```

```python
# Use Groupby() function to calculate the AGe_group wise Descriptive Analysis of a qualitative variable

# Method-1
sales.groupby(['Age_Group'])['Month'].describe()  
```

```python
# Method-2
sales[['Month','Age_Group']].groupby('Age_Group').describe()
```

```python
# Calculate the AGe_group wise Descriptive Analysis of the Unit_Cost and Profit Variables (Quantitative Variables)
sales[['Age_Group','Unit_Cost','Profit']].groupby('Age_Group').describe()
```

```python
em = sales.groupby(['Country', 'Age_Group'])['Country'].count().unstack().plot(kind='bar', stacked=False, figsize=(14,6))
em.set_ylabel('Number of Customers')
em.set_xlabel('Country')
```

```python
sal = sales.groupby(['Country', 'Age_Group']).size().unstack().plot(kind='bar', stacked=True, figsize=(14,6))
sal.set_ylabel('Number of Customers')
```

### 4. Use of Apply Function in Pandas
- Pandas Apply function returns some value after passing each row/column of a data frame with some function. The function can be both default or user-defined.
- For the data manipulation and cleaning we might want to do some complex calculation with the data for that we may need to define functions etc.

```python
# How to write a function in Python?

# squares a given value
def square(x):
    return x ** 2
```

```python
square(2)
```

```python
square(5)
```

```python
# calculates avgerage of two given numbers
def Avg(x, y):
    return (x + y) / 2.0
```

```python
Avg(5,6)
```

```python
df = pd.DataFrame({
    'a': [5, 10 , 15],
    'b': [10 , 20, 30]
})
df
```

```python
# Take the Square of the first column 
df['a'] ** 2
# Similar thing can be down use 'apply' function 
```

```python
# Pass our defined function to 'apply' function 
df['a'].apply(square)
```

```python
# Creat a function that just print nothing else
def print_it(a):
    print(a)
```

```python
df
```

```python
# Here we use the apply function and passes the our own defined function
# So, it will print the values column wise
# In the last part of the output we saw None, None... because our defined function return nothing 
df.apply(print_it)
```

```python
def Avg3(x, y, z):
    return (x + y + z) / 3
```

```python
# Compute the average column wise of the above dataframe 
# ------------------------------------------------------

# When we use the apply function here it will give the error
# because 'apply' function only pick the first column of the data in this way
df.apply(Avg3)
```

```python
# To deal with this issue, we need to define our function as follows:
def Avg3(col):
    x = col[0]
    y = col[1]
    z = col[2]
    return (x + y + z) / 3
```

```python
df.apply(Avg3)
```

```python
# Compute the average row wise of the above dataframe 
# ---------------------------------------------------
def Avg2(row):
    x = row[0]
    y = row[1]
    return (x + y) / 2
df.apply(Avg2, axis=1)
```

### 5. Dealing with the missing values using Pandas

- 'fillna()â€™ does it in one go. 

```python
#First we import scipy function to determine the mode
from scipy.stats import mode
mode(worldmeter['WHO Region'])
```

```python
mode(worldmeter['WHO Region']).mode[0]
```

```python
#Impute the values:
worldmeter['WHO Region'].fillna(mode(worldmeter['WHO Region']).mode[0], inplace=True)
worldmeter['Continent'].fillna(mode(worldmeter['Continent']).mode[0], inplace=True)

#Now check the #missing values again to confirm:
print (worldmeter.apply(missing_number, axis=0))
```

```python
import numpy as np
np.mean(worldmeter['Population'])
```

```python
worldmeter['Population'].fillna(np.mean(worldmeter['Population']), inplace=True)

#Now check the #missing values again to confirm:
print (worldmeter.apply(missing_number, axis=0))
```

```python
# Second method to deal with the missiing values

# Using 'method=ffill' (which is forward fill); means filling missing value with the previous value in the column
# if we check NewCases column 'Nan' replace with the '357.0' 
worldmeter.fillna(method='ffill').head()
```

```python
wmf = worldmeter.fillna(method='ffill').head()
```

```python
print (wmf.apply(missing_number, axis=0))
```

```python
# Third method to deal with the missiing values
wmb = worldmeter.fillna(method='bfill').head()
```

```python
print (wmb.apply(missing_number, axis=0))
```
